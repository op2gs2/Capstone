```python
import numpy as np
import os, array
import imageio # scipy.misc 대신 사용
import glob
from PIL import Image

# DecompressionBombError를 무시
Image.MAX_IMAGE_PIXELS = None

class IMAGE_feature():

    def __init__(self, in_path, out_path):

        self.in_path = in_path
        self.out_path = out_path

    # 
    def get_image(self, path, file):
        # 프로그램이 존재하는 경로를 지정
        filename = path + file

        # 이미지화 할 프로그램을 불러들임
        f = open(filename,'rb')
        ln = os.path.getsize(filename) # 프로그램의 크기

        #
        width = int(ln**0.5) # 파일 길이의 제곱근을 구함(정사각형 모양 이미지 구현을 위해)
        rem = ln % width

        a = array.array("B") # uint8 배열을 선언
        a.fromfile(f,ln-rem) # 파일 객체의 내용을 지정된 수만큼 읽어들임 (파일의 바이너리로 배열을 구성)
        f.close()

        g = np.reshape(a, (int(len(a)/width), width)) # 배열의 모양을 정사각형으로 재조정
        g = np.uint8(g) #g np배열을 uint8형으로 형변환

        # 이미지가 저장될 경로를 지정
        fpng = self.out_path + file + ".png"
        imageio.imwrite(fpng, g) # [수정사항] scipy.misc.imsave(fpng, g) 대신 사용

        outfile = self.out_path + file + "_thumb.png"
        print(outfile)
        size = 256, 256

        if fpng != outfile:
            im = Image.open(fpng)
            im.thumbnail(size, Image.ANTIALIAS) # 이미지를 256x256 크기의 썸네일로 생성
            im.save(outfile, "PNG")

    def get_all(self):
        path = self.in_path

        for file in os.listdir(path): 
            self.get_image(path, file)



def main():

    #mal_path = '/home/user/Desktop/Capstone/dataset/malware/'
    nor_path = '/home/user/Desktop/Capstone/dataset/normal/'
    #nor_path = '/home/user/Desktop/test/'
    #mal_out_path = '/home/user/Desktop/Capstone/dataset/image/malware/'
    nor_out_path = '/home/user/Desktop/Capstone/dataset/image/normal/'
    #nor_out_path = '/home/user/Desktop/test/'
    #im1 = IMAGE_feature(mal_path, mal_out_path)
    #im1.get_all()

    path = r'D:\OneDrive - 한림대학교\바탕 화면\Working Space\Capstone\src\2. 악성코드 분석\'
    im2 = IMAGE_feature(nor_path, nor_out_path)
    im2.get_all()


if __name__ == '__main__':
    main()
```
