```python
import numpy as np
import os, array
import imageio # scipy.misc  
import glob
from PIL import Image

# DecompressionBombError 
Image.MAX_IMAGE_PIXELS = None

class IMAGE_feature():

    def __init__(self, in_path, out_path):

        self.in_path = in_path
        self.out_path = out_path

    # 
    def get_image(self, path, file):
        #    
        filename = path + file

        #    
        f = open(filename,'rb')
        ln = os.path.getsize(filename) #  

        #
        width = int(ln**0.5) #    (    )
        rem = ln % width

        a = array.array("B") # uint8  
        a.fromfile(f,ln-rem) #       (   )
        f.close()

        g = np.reshape(a, (int(len(a)/width), width)) #    
        g = np.uint8(g) #g np uint8 

        #    
        fpng = self.out_path + file + ".png"
        imageio.imwrite(fpng, g) # [] scipy.misc.imsave(fpng, g)  

        outfile = self.out_path + file + "_thumb.png"
        print(outfile)
        size = 256, 256

        if fpng != outfile:
            im = Image.open(fpng)
            im.thumbnail(size, Image.ANTIALIAS) #  256x256   
            im.save(outfile, "PNG")

    def get_all(self):
        path = self.in_path

        for file in os.listdir(path): 
            self.get_image(path, file)



def main():

    #mal_path = '/home/user/Desktop/Capstone/dataset/malware/'
    nor_path = '/home/user/Desktop/Capstone/dataset/normal/'
    #nor_path = '/home/user/Desktop/test/'
    #mal_out_path = '/home/user/Desktop/Capstone/dataset/image/malware/'
    nor_out_path = '/home/user/Desktop/Capstone/dataset/image/normal/'
    #nor_out_path = '/home/user/Desktop/test/'
    #im1 = IMAGE_feature(mal_path, mal_out_path)
    #im1.get_all()

    path = r'D:\OneDrive - \ \Working Space\Capstone\src\2.  \'
    im2 = IMAGE_feature(nor_path, nor_out_path)
    im2.get_all()


if __name__ == '__main__':
    main()
```
