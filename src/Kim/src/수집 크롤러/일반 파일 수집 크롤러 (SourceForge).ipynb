```python
# URL 접속을 위한 라이브러리
import urllib.request
from urllib.request import urlopen

# HTML 파싱을 위한 라이브러리
from bs4 import BeautifulSoup
import re

# 기타 라이브러리
import time as time # for sleep()
from zipfile import ZipFile # .zip 파일 해제
from py7zr import SevenZipFile # .7z 파일 해제
import os
```

```python
def download_from_url(project_names):
    print("***** 정상 파일 수집하기 *****")
    file_count = 0
    file_name = ""
    unknown_file_format = []

    # 링크에 순차적으로 접근하여, 다운로드 받기
    for name in project_names:
        link = "https://sourceforge.net"+str(name[2:-2])+"files/latest/download"

        # 링크에 접속하여, 파일 데이터 받기
        try:
            req = urllib.request.Request(link)
            data = urlopen(req).read()
            time.sleep(1)
        except:
            print("You Gave a Wrong URL. Check it again.")

        # 다운 받은 데이터의 확장자 파악하여, 이름 짓기
        if data[:2] == b"\x4D\x5A":
            file_name = name[12:-3]+".exe"
        elif data[:2] == b'\x50\x4B':
            file_name = name[12:-3]+".zip"
        elif data[:2] == b'\x37\x7A':
            file_name = name[12:-3]+".7z"
        else:
            file_name = name[12:-3] + ".unknown"
            unknown_file_format.append(data[:4])

        # 다운 받은 데이터를 저장하기
        with open("./normal/"+file_name, "wb") as file:
            file.write(data)


        # 다운로드 정보 출력
        file_count += 1
        print(file_count,"번째 파일 다운로드 완료: ",file_name)

print("***** 프로젝트 이름 수집하기 *****")

start_time = time.time()
for page_num in range(1,1000):
    url = "https://sourceforge.net/directory/os:windows/?page="+str(page_num)
    project_names = []

    print(page_num,"번째 페이지 정보 수집중")

    # URL에 접근하기
    try:
        req = urllib.request.Request(url)
        html = urlopen(req).read()
        time.sleep(1)
    except:
        print("You Gave a Wrong URL. Check it again.")

    # BS를 이용해 html 파싱하여, 프로젝트 이름 추출
    project_name_link = '/projects/?.*/'
    name_pattern = re.compile(project_name_link)
    bsObj = BeautifulSoup(html.decode('utf-8'),'html.parser')
    for name in bsObj.findAll('a',class_='result-heading-title'):
        name = name_pattern.findall(str(name))
        project_names.append(str(name))

    # 악성코드를 다운 받는 함수 호출
    download_from_url(project_names)

# 다운로드 시간 출력
end_time = time.time()
exec_time = end_time - start_time

print("\n********** 작업 시각 알림 **********")
print("작업 시작 시각: ",time.ctime(start_time))
print("작업 종료 시각: ",time.ctime(start_time))

exec_time_hour = int(exec_time) // (60*60)
exec_time_min = (int(exec_time) - exec_time_hour * 60*60) // 60
exec_time_sec = (int(exec_time) - exec_time_hour * 60*60 - exec_time_min * 60)
print(exec_time_hour,"시간 ",exec_time_min,"분 ",exec_time_sec,"초 ")
```

```python
dir = './test/'
for file_name in os.listdir(dir):
    print(file_name)
    file = open(dir+file_name,"rb")
    if file.read(2) == b'\x50\x4B':
        print(file_name)
        # zip 파일 해제 및 원본 파일 삭제
        ZipFile(dir+file_name).extractall(path=dir)
        # 압축 파일의 안이 PE 파일이 아닐 경우, 삭제
        for inner_file in os.walk(dir+file_name):
            if inner_file[:2] != b"\x4D\x5A":
                os.remove(inner_file)

    # zip 파일 내 PE 파일은 상위 폴더로 이동, 원본 폴더는 삭제
    if file.read(2) == b'7z':
        print(file_name)
        # .7z 파일 해제 및 원본 파일 삭제
        with SevenZipFile(dir+file_name, 'r') as zf:
            zf.extract(path=dir)
        # 압축 파일의 안이 PE 파일이 아닐 경우, 삭제
        for inner_file in os.walk(file_name):
            if inner_file[:2] != b"\x4D\x5A":
                os.remove(inner_file)
```
