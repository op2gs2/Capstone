```python
import pefile

def NT_Header_File_Header(pe):
    print("NumberOfSections: ",pe.FILE_HEADER.NumberOfSections)
    print("Machine :",hex(pe.FILE_HEADER.Machine))
```

```python
def NT_Header_Optional_Header(pe):
    print("AddressOfEntryPoint: ",hex(pe.OPTIONAL_HEADER.AddressOfEntryPoint))
    print("ImageBase: ",hex(pe.OPTIONAL_HEADER.ImageBase))
    print("SectionAlignment: ",hex(pe.OPTIONAL_HEADER.SectionAlignment))
    print("FileAlignment: ",hex(pe.OPTIONAL_HEADER.FileAlignment))
    print("SizeOfImage: ",hex(pe.OPTIONAL_HEADER.SizeOfImage))
    print("SizeOfHeader: ",hex(pe.OPTIONAL_HEADER.SizeOfHeaders))
```

```python
def Section_Header_Extraction(pe):
    for section in pe.sections:
        print("섹션 이름: ", section.Name, "\nVirtualSize: ", hex(section.Misc_VirtualSize),
              "\nVirtualAddress: ",hex(section.VirtualAddress), "\nSizeOfRawData: ",hex(section.SizeOfRawData),"\nPointerToRawData :",hex(section.PointerToRawData),"\n")
```

```python
def import_apis(pe):
    pe.parse_data_directories()

    try:
        for entry in pe.DIRECTORY_ENTRY_IMPORT:
          print(entry.dll)
          for imp in entry.imports:
            print('\t', hex(imp.address), imp.name)
    except:
        print("Import 항목이 없습니다.")
        pass
```

```python
def export_apis(pe):
    try:
        for exp in pe.DIRECTORY_ENTRY_EXPORT.symbols:
            print(hex(pe.OPTIONAL_HEADER.ImageBase + exp.address), exp.name, exp.ordinal)
    except:
        print("Export 항목이 없습니다.")
        pass
```

```python
pe = pefile.PE('./python-3.10.3-amd64.exe')

print("NT Header - File Headers\n")
NT_Header_File_Header(pe)
print("\nNT Header - Optional Headers\n")
NT_Header_Optional_Header(pe)
print("\nSection Header\n")
Section_Header_Extraction(pe)
print("\nImport APIs\n")
import_apis(pe)
print("\nExport APIs\n")
export_apis(pe)
```
